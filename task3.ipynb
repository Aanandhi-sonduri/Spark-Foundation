```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Heatmap

```python
sns.heatmap(df.corr(),annot=True)
plt.figure(figsize=(5,1))
plt.show()
```

# Pairplot

```python
sns.pairplot(df,hue='Region')
```

# Bargraph

```python

plt.figure(figsize=(15,10))
df['State'].value_counts().plot.bar()
plt.xlabel("States",size=14)
plt.xticks(size=14)
plt.ylabel("Number of orders",size=10)
plt.show()
```

# Pieplot 

```python
plt.figure(figsize=(18,18))
df.groupby("State").sum()["Sales"].plot.pie(autopct="%1.1f%%")
plt.title('State_wise analysis of Sales',fontsize=30)
```

```python
plt.figure(figsize=(10,10))
df.groupby("Region").sum()["Profit"].plot.pie(autopct="%1.1f%%")
plt.title('Region-wise analysis of Profit',fontsize=20)
```

```python
plt.figure(figsize=(10,10))
df.groupby("Category").sum()["Profit"].plot.pie(autopct="%1.1f%%")
plt.title('Category-wise analysis of Profit',fontsize=20)
```
